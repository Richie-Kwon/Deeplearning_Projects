```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime

df = pd.read_csv(r'C:\Users\RICHARDSON\Desktop\Python\tsa-tutorial-master\tsa-tutorial-master\data\apple_stock.csv')
df['Date'] = df['Date'].apply(pd.to_datetime)

df.set_index('Date', inplace=True)
df['month'] = df.index.month
df.groupby('month').agg(sum)
```

```python
df.groupby(df.index.year).sum()
```

```python
df.groupby([df.index.year, df.index.month]).agg(sum)
```

```python
df.resample(rule='A').sum()
df.resample(rule='A').mean()

df.resample(rule='A').mean()['2009']

def first_day(sample):
    return sample[0]
    
df.resample(rule='A').apply(first_day)
```

```python
def dateparse(str_dt):
    return pd.datetime.strptime(str_dt, '%Y-%M-%D')
    
df = pd.read_csv(r'C:\Users\RICHARDSON\Desktop\Python\tsa-tutorial-master\tsa-tutorial-master\data\apple_stock.csv', 
                 index_col ='Date', 
                 parse_dates= ['Date'],
                 date_parser= dateparser)
```

```python
df.rolling(7).mean().head() #7일단위로 묶어서 평균내기
```

```python
df['Close'].plot()
df.rolling(30).mean()['Close'].plot()
```

```python
df['2018':]['Close'].plot()
df['2018':].rolling(window = 7).mean()['Close'].plot()
```

```python
df['Close: 30 Day mean'] = df['Close'].rolling(30).mean()
df['Close: 30 Day mean'].plot(figsize = (15,7))
```

```python
df['Close'].expanding(min_periods=1).mean().plot(figsize = (15,7))
```

```python
df['Close 20 days rolling'] = df['Close'].rolling(20).mean()
df['Upper'] = df['Close 20 days rolling'] + 2 * df['Close'].rolling(20).std()
df['Lower'] = df['Close 20 days rolling'] - 2 * df['Close'].rolling(20).std()

df['2018':][['Close', 'Close 20 days rolling','Upper','Lower']].plot(figsize=(15,7))
```
